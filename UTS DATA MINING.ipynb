```python
import numpy as np
import pandas as pd
```

```python
temperatures = [25, 28, 30, 22, 35, 40, 18, 24]

# Konversi ke Fahrenheit
temp_fahrenheit = [(c * 9/5) + 32 for c in temperatures]
print("Suhu dalam Fahrenheit:", temp_fahrenheit)

# Slicing 3 data terakhir
last_three = temp_fahrenheit[-3:]
print("3 data terakhir:", last_three)
```

```python
inventory = {
    "laptop": {"stok": 10, "harga": 15000000},
    "mouse": {"stok": 50, "harga": 250000},
    "monitor": {"stok": 20, "harga": 3000000}
}

# Tambah keyboard
inventory["keyboard"] = {"stok": 30, "harga": 500000}

# Update harga laptop
inventory["laptop"].update({"harga": 14500000})

# Hitung total aset
total_aset = sum(item["stok"] * item["harga"] for item in inventory.values())
print(f"Total nilai aset: Rp{total_aset:,}")
```

```python
scores = np.array([
    [80, 85, 90],
    [70, 60, 75],
    [95, 90, 100],
    [40, 50, 45]
])

rata_rata = np.mean(scores, axis=1)
nilai_tertinggi = np.max(scores)
status = np.where(rata_rata >= 70, "Lulus", "Gagal")

print("Rata-rata:", rata_rata)
print("Nilai Tertinggi:", nilai_tertinggi)
print("Status:", status)
```

```python
df_data = {
    'Nama': ['Andi', 'Budi', 'Caca', 'Dedi', 'Euis'],
    'Departemen': ['IT', 'HR', 'IT', 'Sales', 'Sales'],
    'Gaji': [8000000, 7000000, 8500000, np.nan, 6000000],
    'Pengalaman_Tahun': [5, 3, 6, 2, 1]
}

df = pd.DataFrame(df_data)

# Isi NaN dengan median
df['Gaji'] = df['Gaji'].fillna(df['Gaji'].median())

# Kategori Senioritas
df['Kategori_Senioritas'] = np.where(df['Pengalaman_Tahun'] > 4, 'Senior', 'Junior')

# Rata-rata gaji per departemen
rata_gaji = df.groupby('Departemen')['Gaji'].mean()

print(df)
print("\nRata-rata per Departemen:\n", rata_gaji)
```
